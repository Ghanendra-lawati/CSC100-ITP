### Starter Code: Word Scrambler

```python
print("=== Word Scrambler Game ===\n")

# Word list (add more!)
words = ["python", "program", "computer", "science", "coding"]

# Game variables
score = 0
round_num = 1
playing = True

while playing and round_num <= len(words):
    # Get word for this round
    original = words[round_num - 1]
    
    # Simple scramble (reverse for now)
    scrambled = original[::-1]
    
    print(f"\nRound {round_num}")
    print(f"Scrambled word: {scrambled}")
    print(f"Hint: {len(original)} letters, starts with '{original[0]}'")
    
    # Get guess
    attempts = 3
    guessed = False
    
    while attempts > 0 and not guessed:
        guess = input(f"\nYour guess ({attempts} left): ").strip().lower()
        
        if guess == original:
            print("Correct! +10 points")
            score = score + 10
            guessed = True
        else:
            attempts = attempts - 1
            if attempts > 0:
                print("Wrong! Try again.")
                # Give another hint
                if attempts == 1:
                    print(f"Hint: ends with '{original[-1]}'")
    
    if not guessed:
        print(f"Out of attempts! Word was: {original}")
    
    # Next round?
    round_num = round_num + 1
    if round_num > len(words):
        print(f"\nGame Over! Final score: {score}")
        playing = False
    else:
        continue_game = input("\nContinue? (y/n): ").lower()
        if continue_game != 'y':
            playing = False

print(f"\nThanks for playing! Final score: {score}")
```

# Common Issues and Solutions

## Slicing Errors

**Problem:**
```python
text = "Hello"
last_two = text[-2]  # Gets one character!
```

**Solution:** Remember slicing needs start:end
```python
last_two = text[-2:]  # Gets "lo"
```

## String Immutability

**Problem:**
```python
word = "test"
word[0] = "T"  # Error!
```

**Solution:** Build new string
```python
word = "T" + word[1:]  # "Test"
```

## Split Edge Cases

**Problem:**
```python
"a,,b".split(",")  # ['a', '', 'b'] - empty string!
```

**Solution:** Filter empty strings
```python
parts = [p for p in text.split(",") if p]  # Advanced!
# Or with loop:
clean_parts = []
for part in text.split(","):
    if part:  # Not empty
        clean_parts.append(part)
```

## Method Order Matters

**Problem:**
```python
"  HELLO  ".lower().strip()  # "  hello  "
```

**Solution:** Strip first, then transform
```python
"  HELLO  ".strip().lower()  # "hello"
```
